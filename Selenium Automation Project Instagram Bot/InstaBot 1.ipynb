## 1.LogIn 

### Submit with sample username and password

```python
def login(username,password):
    try:
        driver.find_element(By.XPATH,"//*[@id='loginForm']/div/div[1]/div/label/input").send_keys(username)
        driver.find_element(By.XPATH,"//*[@id='loginForm']/div/div[2]/div/label/input").send_keys(password)
        driver.find_element(By.XPATH,"//*[@id='loginForm']/div/div[3]").click()
        #Going through popups
        pop = wait.until(EC.presence_of_element_located((By.XPATH,".//div[(@class='_ac8f')]/div")))
        pop.click()
        #closing the notification popup
        notify = wait.until(EC.presence_of_element_located((By.XPATH,".//div[(@class='_a9-z')]/button[2]")))
        notify.click()
    except TimeoutException:
        print("Something is Wrong")
```

### 2. This Search option Script will help you get all the profile name with specific keywords.

##### Note : Make sure to avoid printing hashtags

```python
def search(s):
    try:
        #opening search box for search operation
        wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,"Search"))).click()
        search = wait.until(EC.presence_of_element_located((By.XPATH,".//div[(@class ='xjoudau x6s0dn4 x78zum5 xdt5ytf x1c4vz4f xs83m0k xrf2nzk x1n2onr6 xh8yej3 x1hq5gj4')]/input")))
        search.send_keys(s)
        #extracting usernames
        wait.until(EC.presence_of_element_located((By.XPATH,".//div[(@class = 'x9f619 x1n2onr6 x1ja2u2z x78zum5 x1iyjqo2 xs83m0k xeuugli x1qughib x6s0dn4 x1a02dak x1q0g3np xdl72j9')]/div/div/div/span")))
        handel_name = driver.find_elements(By.XPATH,".//div[(@class = 'x9f619 x1n2onr6 x1ja2u2z x78zum5 x1iyjqo2 xs83m0k xeuugli x1qughib x6s0dn4 x1a02dak x1q0g3np xdl72j9')]/div/div/div/span")
        names = []
        for i in handel_name:
            if i.text[0] != "#":
                names.append(i.text.split('\n')[0])
        time.sleep(3)
        driver.refresh()
        return names
    except TimeoutException:
        print("Something is Wrong")
```

## 3.Searching and Opening A Profile 

```python
def search_open(s):
    try:
        #locating Search Box and Sending text for the search
        wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,"Search"))).click()
        search = wait.until(EC.presence_of_element_located((By.XPATH,".//div[(@class ='xjoudau x6s0dn4 x78zum5 xdt5ytf x1c4vz4f xs83m0k xrf2nzk x1n2onr6 xh8yej3 x1hq5gj4')]/input")))
        search.send_keys(s)
        
        #Checking the result
        result = wait.until(EC.presence_of_element_located((By.XPATH,".//div[@class = 'x9f619 x78zum5 xdt5ytf x1iyjqo2 x6ikm8r x1odjw0f xh8yej3 xocp1fn']/a")))
        result.click()
        
        time.sleep(5)
    except TimeoutException:
        print('Something Wrong with the search and open OP')
```

## 4 : Follow/Unfollow given handle :-

```python
def follow():
    try:
        
        #Locating follow button
        follow_btn = wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[2]/div/div[2]/section/main/div/header/section[2]/div/div/div[2]/div/div[1]/button")))
        
        #checking for the handle follow followed by a message for follow
        if follow_btn.text == 'Follow':
            follow_btn.click()
            time.sleep(3)
        else:
            print("Already Following")
    except TimeoutException:
        print("something Wrong with the Follow OP")
```

```python
def unfollow():
    try:
        follow_btn = wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[2]/div/div[2]/section/main/div/header/section[2]/div/div/div[2]/div/div[1]/button")))
        if follow_btn.text != 'Follow':
            follow_btn.click()
            time.sleep(3)
            #going Through the popup for conformation to unfollow
            pop_up = wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[6]/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div/div[8]/div[1]")))
            pop_up.click()
            time.sleep(3)
        else:
            print("Already Unfollowed")
    except TimeoutException:
            print("Something Wrong with the Unfolow OP")
```

## 5 : Like/Unlike posts 

####  1. Liking the top n number of  posts in a profile.

```python
def like_post():
    try:
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
        time.sleep(2)
        driver.execute_script('window.scrollTo(0,document.body.scrollTop);')
        time.sleep(2)
        post = driver.find_elements(By.CLASS_NAME,"_aagu")
        for i in range(30):
            time.sleep(2)
            post[i].click()
            
            #locating the link button/ Unlike button
            wait.until(EC.presence_of_element_located((By.XPATH,"//section/span")))
            like_btn = driver.find_element(By.XPATH,"//section/span")
            status = bs(like_btn.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            if  status == 'Like':
                like_btn.click()
                like_btn.click()
                time.sleep(2)
                print("Post",i," is Liked")
            else:
                print("You Have already Liked the post",i+1)
                time.sleep(1)
                #locating the 'X' button to close the post.
            wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[8]/div[1]/div/div[2]"))).click()
    except TimeoutException:
        print("Something Is wrong with Like OP")
```

```python
def unlike_post():
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        time.sleep(3)
        driver.execute_script("window.scrollTo(0, document.body.scrollTop);")
        time.sleep(2)
        post = driver.find_elements(By.CLASS_NAME,"_aagu")
        
        for i in range(30):
            time.sleep(2)
            post[i].click()
            
            
            #locating the Like/Dislike Button For Further OP
            wait.until(EC.presence_of_element_located((By.XPATH,"//section/span")))
            like_btn = driver.find_element(By.XPATH,"//section/span")
            status = bs(like_btn.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            
            if status == 'Unlike':
                like_btn.click()
                time.sleep(2)
                print("Post",i," is UnLiked")
                
            else:
                print("You Have Already UnLiked The Post",i+1)
                time.sleep(2)
            #locating the close button
            wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[8]/div[1]/div/div[2]"))).click()
    except TimeoutException:
        print("Something Wrong with the Dislike OP")
```

##  6 : Extract list of followers

1.Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
2.Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def extract_followers():
    try:
        #locating the followers button for getting the followers name
        
        followers_btn = wait.until(EC.element_to_be_clickable((By.XPATH,"/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[2]/div/div[2]/section/main/div/header/section[3]/ul/li[2]/div/a")))
        driver.execute_script("arguments[0].scrollIntoView(true);", followers_btn)
        time.sleep(1)
        followers_btn.click()
            
            
        #followers list frame
        try:
            frame = wait.until(EC.presence_of_element_located((By.XPATH, ".//div[@class = 'xyi19xy x1ccrb07 xtf3nb5 x1pc53ja x1lliihq x1iyjqo2 xs83m0k xz65tgg x1rife3k x1n2onr6']")))
        except StaleElementReferenceException:
            print("StaleElementReferenceException: Re-locating the frame.")
            frame = wait.until(EC.presence_of_element_located((By.XPATH, ".//div[@class = 'xyi19xy x1ccrb07 xtf3nb5 x1pc53ja x1lliihq x1iyjqo2 xs83m0k xz65tgg x1rife3k x1n2onr6']")))
        for i in range(50):
            driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", frame)
            time.sleep(2)
        
        names = []
        
        #extracting the Names of the followers
        followers_name = wait.until(EC.presence_of_all_elements_located((By.XPATH,"//span[@class='_ap3a _aaco _aacw _aacx _aad7 _aade']")))
        
        for i in followers_name[:500]:
            names.append(i.text.split('\n')[0])
        return names
    except TimeoutException:
        print("Something Wrong with the OP")
    except Exception as e:
        print(f"An error occurred: {e}")
```
