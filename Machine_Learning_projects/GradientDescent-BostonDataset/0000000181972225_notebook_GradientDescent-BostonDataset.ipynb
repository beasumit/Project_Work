```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

data=np.genfromtxt(r"C:\Users\bensu\Downloads\0000000000002417_training_boston_x_y_train (1).csv",delimiter =',')
test_data = np.genfromtxt(r"C:\Users\bensu\Downloads\0000000000002417_test_boston_x_test (1).csv",delimiter =',')

def step_gradient(points,learning_rate,M):
    N=points.shape[0]
    num_col = points.shape[1]
    new_M=np.zeros(num_col)
    for i in range(N):
        x = points[i, 0:num_col-1]
        x = np.append(x, 1)
        y=points[i,num_col-1]
        for j in range(num_col):
            new_M[j] += (-2/N) * (y - (M * x).sum()) * x[j]
        M = M - (learning_rate * new_M)
    return M
    
def gd(points, learning_rate, num_iteration):
    num_col = points.shape[1]
    M = np.zeros(num_col)
    
    for i in range(num_iteration):
        M = step_gradient(points, learning_rate, M)
        print(i, "Cost= ", cost(points, M))
    return M

def cost(points, M):
    total_cost = 0
    
    N = points.shape[0]
    num_col = points.shape[1]
    
    for i in range(N):
        x = points[i,0:num_col-1]
        x = np.append(x, 1)
        y = points[i,num_col-1]
        total_cost +=(1/N) * (y - (M * x).sum()) ** 2
        
    
    return total_cost

def run(data):
    data = data
    learning_rate=0.001
    num_iteration=1000
    #m,c=gd(data,learnning_rate,num_iteration)
    m=gd(data,learning_rate,num_iteration)
    return m[0:13],m[13]
```

```python
m,c= run(data)
print(m,c)
```

```python
x=data[:, :-1]
y=data[:, -1]
scaler= StandardScaler()
scaler.fit(x)
x=scaler.transform(x)

test_trans=scaler.transform(test_data)
def predict(x, m, c):
    N = x.shape[0]
    y = np.zeros(N)
    
    for i in range(N):
        X = x[i,:]
        y[i] = (m * X).sum() + c
        
    return y
```
